```python
sql("""
--CREATE TEMP TABLE call_map AS SELECT * FROM cregit_calls WHERE file LIKE 'drivers/sh/intc/%';
CREATE TEMP TABLE call_map AS SELECT file, 
    FROM cregit_identifiers WHERE file LIKE 'drivers/sh/intc/%';
    
WITH RECURSIVE dependencies(file, caller, callee, indicator) AS (
    SELECT file, caller, callee, 'direct' FROM call_map
    UNION
    SELECT call_map.file, call_map.caller, dependencies.callee, 'transitive' AS callee
    FROM call_map JOIN dependencies ON
        call_map.callee = dependencies.caller AND call_map.file LIKE 'drivers/sh/intc/access.c'
)
SELECT * FROM dependencies
;""")
```
